```python
import pandas as pd
import numpy as na
from sklearn import linear_model
import matplotlib.pyplot as plt
url='https://raw.githubusercontent.com/apratim777/apratim777/master/homeprices.csv'
df=pd.read_csv(url)
df
```

```python
df.tail(n=2)
df.head(n=2)
```

```python
plt.xlabel('Area')
plt.ylabel('Price')
plt.scatter(df.area,df.price,color ='green',marker='+')
plt.show()
```

```python
reg=linear_model.LinearRegression()
new_x=df.drop('price',axis='columns')
new_x
```

```python
reg.fit(new_x,df.price)
```

```python
reg.predict([[3200]])
```
