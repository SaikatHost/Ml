```python
import pandas as pd
import numpy as na
from sklearn import linear_model
import matplotlib.pyplot as plt
url='https://raw.githubusercontent.com/apratim777/apratim777/master/homeprices1.csv'
#url='https://raw.githubusercontent.com/apratim777/apratim777/master/homeprices2.csv'
df=pd.read_csv(url)
df
```

```python
mid=df.bedrooms.median()
mid
```

```python
df.bedrooms=df.bedrooms.fillna(mid)
df
```

```python
reg=linear_model.LinearRegression();
reg.fit(df.drop('price',axis='columns'),df.price)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[3000,4.0,15]])
```
